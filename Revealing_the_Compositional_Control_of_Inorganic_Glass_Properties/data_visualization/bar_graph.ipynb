```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np
from new_plot import *
set_things()
import json
```

```python
df = pd.read_csv('prop_rows_cols.csv')
```

```python
set_font(size=15, family='Arial', weight='normal')
```

```python
cols = df.columns
rowsdf = pd.DataFrame(columns=cols)
for cat in df['CLASS'].unique():
    tempDF = df[df['CLASS']==cat].sort_values(by=['ROWS'],ascending=False)
    rowsdf = rowsdf.append(tempDF)
#     break
```

```python
cols = df.columns
colsdf = pd.DataFrame(columns=cols)
for cat in df['CLASS'].unique():
    tempDF = df[df['CLASS']==cat].sort_values(by=['COLS'],ascending=False)
    colsdf = colsdf.append(tempDF)
#     break
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import gridspec
set_mood_light()
# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig = plt.figure()
# set height ratios for sublots
gs = gridspec.GridSpec(2, 1, height_ratios=[1, 1]) 

# the fisrt subplot
ax0 = plt.subplot(gs[0])
# log scale for axis Y of the first subplot
ax0.set_yscale("log")
line0 = ax0.bar(rowsdf['NAME'].values, rowsdf['ROWS'].values)
ylabel('No. of glasses')
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='minor',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False) 
#the second subplot
# shared axis X
ax1 = plt.subplot(gs[1], sharex = ax0)
line1 = ax1.bar(rowsdf['NAME'].values, rowsdf['COLS'].values)
plt.setp(ax0.get_xticklabels(), visible=False)
# remove last tick label for the second subplot
yticks = ax1.yaxis.get_major_ticks()
yticks[-1].label1.set_visible(False)
# put lened on first subplot
# ax0.legend((line0, line1), ('red line', 'blue line'), loc='lower left')
plt.xticks(rotation='vertical')
plt.ylim([0,230])
# remove vertical gap between subplots
plt.subplots_adjust(hspace=0.)
fig = plt.gcf()
fig.set_size_inches(10,10)
ylabel('No. of components')
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='minor',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False)        # ticks along the top edge are off

plt.savefig('bar_chart.png',dpi=1200,bbox_inches='tight',transparent=False)
plt.show()
```
